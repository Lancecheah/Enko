# Simple Moving Average (SMA)

| -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |

It is calculated by taking a series of prices (or reporting periods), adding these together and then dividing the total by the number of data points.

## Formula (SMA = $\frac{A_1+A_2+...A_n}{n}$)

where:  
${A_n}$ = the price of an asset at period n  
n = the number of total periods

```python
# Declaring Simple Moving Average Perimeters
sma_s = 50 #Short movning average of 50 rolling windows (mean)
sma_l = 200 #Long moving average of 2000 rolling windoes (mean)
```

```python
dfret = pd.DataFrame()
for currency in close.columns:
    dfSMA = close.loc[:, currency].to_frame()
    dfSMA["SMA_S"] = dfSMA[currency].rolling(sma_s).mean()
    dfSMA["SMA_L"] = dfSMA[currency].rolling(sma_l).mean()
    dfSMA.dropna(inplace = True)
    dfSMA["Position"] = np.where(dfSMA["SMA_S"] > dfSMA["SMA_L"], 1, -1)
    dfSMA["returns"] = np.log(dfSMA[currency].div(dfSMA[currency].shift(1)))
    dfSMA["strategy"] = dfSMA.Position.shift(1) * dfSMA["returns"]
    dfSMA.dropna(inplace = True)
    dfSMA["creturns"] = dfSMA["returns"].cumsum().apply(np.exp)
    dfSMA["cstrategy"] = dfSMA["strategy"].cumsum().apply(np.exp)
    dfret[currency] = dfSMA["cstrategy"] #for ploting return of investment later
    dfSMA[["creturns", "cstrategy"]].plot(figsize = (12, 8), title = "{} - SMA{} | SMA{}".format(currency, sma_s, sma_l), fontsize = 12)
    plt.legend(fontsize = 12)
    plt.show()
    strategy = dfSMA["cstrategy"].iloc[-1]
    ret =strategy - dfSMA["creturns"].iloc[-1]
    print(f" SMA Strategy = {strategy:.4f} & differences = {ret:.4f}")
```

```python
dfret = dfret.replace(np.nan, 1.0)
dfret = dfret*1000
```

```python
dfret.plot(figsize = (15,8), fontsize = 13)
plt.legend(fontsize=13)
plt.xlabel("Dates", fontsize = 15)
plt.ylabel("Return of Investment", fontsize = 15)
plt.title("Initial $1,000 Investment for SMA", fontsize = 20)
plt.show()
```

# Bollinger Bands

| -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |

A Bollinger Band is a technical analysis tool defined by a set of trendlines plotted two standard deviations (positively and negatively) away from a simple moving average (SMA) of a security's price.

## Formula (BOLU=MA(TP,n)+m∗σ[TP,n])

where:  
BOLU=Upper Bollinger Band  
BOLD=Lower Bollinger Band  
MA=Moving average  
TP (typical price)=(High+Low+Close)÷3  
n=Number of days in smoothing period (typically 20)  
m=Number of standard deviations (typically 2)  
σ[TP,n]=Standard Deviation over last n periods of TP



```python
SMA = 30
dev = 2
```

```python
norm_df1= pd.DataFrame()
for currency in close.columns:    
    dfbb = close.loc[:, currency].to_frame()
    dfbb["returns"] = np.log(dfbb.div(dfbb.shift(1)))
    dfbb["SMA"] = dfbb[currency].rolling(SMA).mean()
    dfbb["Lower"] = dfbb["SMA"] - dfbb[currency].rolling(SMA).std() * dev
    dfbb["Upper"] = dfbb["SMA"] + dfbb[currency].rolling(SMA).std() * dev
    dfbb.dropna(inplace = True)
    dfbb["distance"] = dfbb[currency] - dfbb.SMA #To measure the distance between the price nearer to lower or Upper
    dfbb["position"] = np.where(dfbb[currency] < dfbb.Lower, 1, np.nan) # Buy
    dfbb["position"] = np.where(dfbb[currency] > dfbb.Upper, -1, dfbb["position"]) # Sell
    dfbb["position"] = np.where(dfbb.distance * dfbb.distance.shift(1) < 0, 0, dfbb["position"])
    dfbb["position"] = dfbb.position.ffill().fillna(0)
    dfbb["strategy"] = dfbb.position.shift(1) * dfbb["returns"]
    dfbb.dropna(inplace = True)
    dfbb["creturns"] = dfbb["returns"].cumsum().apply(np.exp)
    dfbb["cstrategy"] = dfbb["strategy"].cumsum().apply(np.exp)
    norm_df1[currency] = dfbb["cstrategy"]
    dfbb[["creturns", "cstrategy"]].plot(title =currency, figsize =(12,8))
    plt.show()
    strategy = dfbb["cstrategy"].iloc[-1]
    ret =strategy - dfbb["creturns"].iloc[-1]
    print(f" BOLLINGER BANDS - Strategy = {strategy:.4f} & differences = {ret:.4f}")
```

```python
norm_df1 = norm_df1.replace(np.nan, 1.0)
norm_df1 = norm_df1*1000
```

```python
norm_df1.plot(figsize = (15,8), fontsize = 13)
plt.legend(fontsize=13)
plt.xlabel("Dates", fontsize = 15)
plt.ylabel("Return of Investment", fontsize = 15)
plt.title("Initial $1,000 Investment for Bollinger Bands", fontsize = 20)
plt.show()
```

# Non Payroll Trading Strategy

## If Wednesday > Thursday than Sell else Buy 

```python
final_df = pd.DataFrame()
norm_df = pd.DataFrame()
from pandas.tseries.offsets import WeekOfMonth
for currency in close.columns: 
    nfp = close.loc[:, currency].to_frame()
    wed = nfp.resample(WeekOfMonth(week=0, weekday=2)).asfreq().dropna()
    wed.reset_index(drop=True,inplace=True)
    wed.rename(columns = {currency:'WED'},inplace=True)
    thur = nfp.resample(WeekOfMonth(week=0, weekday=3)).asfreq().dropna()
    thur.reset_index(drop=True,inplace=True)
    thur.rename(columns = {currency:'THUR'},inplace=True)
    fri = nfp.resample(WeekOfMonth(week=0, weekday=4)).asfreq().dropna()
    fri.reset_index(drop=True,inplace=True)
    fri.rename(columns = {currency:'FRI'},inplace=True)
    new_df = pd.concat([wed,thur,fri], axis = 1)
    new_df.dropna(inplace = True)
    new_df["returns"] = new_df["THUR"] - new_df["FRI"]
    new_df["position"] = np.where(new_df["WED"] > new_df["THUR"], -1, 1)
    new_df["strategy"] = new_df.position * new_df["returns"]
    new_df["cstrategy"] = new_df["strategy"].cumsum()
    final_df[currency] = new_df["strategy"].to_frame()
    norm_df[currency] = new_df["cstrategy"].to_frame()
```

```python
print(final_df.to_string())
```

```python
hit_df = pd.DataFrame()
for currency in close.columns: 
    hits = np.sign(final_df[currency]).value_counts()
    hit_df[currency] = hits.to_frame()
```

```python
hits_df = (hit_df.fillna(0).T)
```

```python
hits_df.columns = ["Win", "Lose"]
```

```python
hits_df.plot(kind='bar', figsize = (12,8))
plt.xlabel("Forex Pair", fontsize = 15)
plt.ylabel("Number of Trades", fontsize = 15)
plt.title("Wed > Thur than Sell else Buy", fontsize = 20)
plt.show()
```

# Using Volume Profile to Draw Support & Resistance 

```python
close = close[["AUDUSD"]]
close.reset_index(inplace=True)
```

```python
close = close.rename(columns={'AUDUSD': 'Close'})
close
```

```python
close.columns = ['Date', 'Close']
close['Date'] = pd.to_datetime(close['Date'])
close = close.sort_values('Date', ascending=True)
close = close.dropna() 
close = close[close['Close'] != '.']
close['Close'] = close['Close'].astype(float)

close 
```

```python
# scale values and cast to ints
prices = close['Close'].values
x = list(prices)
prices = ((x-min(x))/(max(x)-min(x))) * 100 + 1
prices
```

```python
values = []
counters = []
for cl in prices:
    # build price and index
    cl = int(cl)
    values.append(cl)
    counters.append(values.count(cl))

profile_df = pd.DataFrame({'Value':values, 'Index':counters})
profile_df.tail()
```

```python
# add the scaled prices back
close['Close_Scaled'] = prices

# bring it all together with chart
fig, ax = plt.subplots(figsize=(16, 8))
plt.plot(close['Date'], close['Close_Scaled'], linewidth=2)
plt.grid()
plt.title('AUDUSD')

ax.twiny()
plt.scatter(profile_df['Index'], profile_df['Value'], s=3, color='black')

plt.show()
```

# Conclusion  
Volume profile accurately show area of the price reverse which is a better way of drawing Support & resistance level.
